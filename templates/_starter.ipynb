```python
from manywidgets import Chart, Slider, Column, Row
from ipywidgets import jsdlink
import numpy as np

x = np.linspace(0, 10, 200)
chart = Chart(title="Hello", x_label="x", y_label="y", height=320, width=520)
chart.add_series(x=x, y=np.sin(x), name="sin")

freq = Slider(label="Frequency", min=1, max=5, value=1)
Row(Column(freq), chart, gap="24px", align="start")
```
